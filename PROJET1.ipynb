###### ETAPE 1 IMPORTATION DU DATA FRAME , DU DATA SET ET AFFICHAGE DU NOMBRES DE VALEURS MANQUANTES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

df = pd.read_csv("sensor_dataset.csv")
print(f"Le nombres de valeurs manquantes du dataset par colonne est : {df.isna().sum()}")
```

###### ETAPE 2 NETOYAGE DU DATASET ET REAFFICHAGE DU NOMBRE DE VALEUR MANQUANTES

```python
df = df.round(2)
df['temperature'] = df['temperature'].fillna(df['temperature'].mean()).round(2)
df['humidity'] = df['humidity'].fillna(df['humidity'].mean()).round(2)
df['pressure'] = df['pressure'].fillna(df['pressure'].mean()).round(2)
print(f"Le nombres de valeurs manquantes du dataset par colonne apres netoyage est : {df.isna().sum()}")
```

###### ETAPE 3 ECRITURES DES FONCTIONS STATISTIQUES PERSONNELLES 

```python
def moyenne(colone):
    return sum(colone)/len(colone)

def nbre_ligne(colone):
    return len(colone)

def minimum(colone):
    mini = 10000000000
    for i in colone:
        if i < mini:
            mini = i
    return mini

def maximum(colone):
    maxi = -1
    for i in colone:
        if i > maxi:
            maxi = i
    return maxi

def modt(colone):
    freq,maxi = {},-1
    for i in colone:
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1
    for i in freq :
        if freq[i] > maxi:
            maxi = freq[i]
            x = i
    return(f"{i} : {maxi} occurence")

def mediane(colone):
    colone = sorted(colone)
    n = len(colone)
    if n % 2== 0:
        return colone[n//2]
    else:
        return colone[(n-1)//2]
    
def variance(colone):
    moy = sum(colone)/len(colone)
    return sum((x-moy)**2 for x in colone)/len(colone)

def ecartType(colone):
    return math.sqrt(variance(colone))

```

###### ETAPE 4 STOCKAGES DES STATISTIQUES DANS UN DICTIONNAIRE

```python
statistiques  = {}
for i in ["temperature","humidity","pressure"]:
    statistiques[i] = {
        "moyenne": moyenne(df[i]),
        "minimum": minimum(df[i]),
        "maximum": maximum(df[i]),
        "variance": variance(df[i]),
        "ecart Type": ecartType(df[i]),
        "mediane": mediane(df[i])
    }
    
```

###### AFFICHAGE DES STATISQUES

```python
for i,j in statistiques.items():
    print(f"\n********** affichage des statistiques concernant : {i} **********")
    for cle , valeur in j.items():
        print(f"{cle}:{valeur}")
```

###### ETAPE 5 VIZUALISATION DES DIFFERENTES COURBES 

```python
def histogramme():
    for i in ["temperature","humidity","pressure"]:
        plt.hist(df[i],bins=20)
        plt.xlabel(f'{i}')
        plt.ylabel("Frequence")
        plt.title(f"Courbe des histogramme des {i}")
        print(f"VOICI L'HISTOGRAMME DE {i}")
        plt.show()

def courbeTemporelle():
    for i in ["temperature","humidity","pressure"]:
        plt.figure(figsize=(20,10))
        plt.plot(df['timestamp'],df[i])
        plt.xlabel("Timestamp")
        plt.ylabel(f'{i}')
        plt.title(f"Courbe temporelle des {i}")
        print(f"VOICI LA COURBE TEMPORELLE DE {i}")
        print(f"Cette courbe represente l'evolution des {i} dans le temps")
        plt.show()

def boiteAMoustache():
    for i in ["temperature","humidity","pressure"]:
        sns.boxplot(x=df[i])
        plt.title(f"BOITES A MOUSTACHES DES  {i}")
        plt.show()
        
def heatMap():
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    sns.heatmap(df.corr(),annot=True, cmap = "coolwarm")
    plt.title("Correlation entre les colones du dataset")
    plt.show()
while (True):
        
    print("""
        Quelles courbes veut tu voir :
            1-HISTOGRAMME
            2-COURBE TEMPORELLE
            3-BOITE A MOUSTACHE
            4-HEADMAP
            Q/q POUR QUITTER
        """)
    choix = str(input("Entre ton choix: "))
    match choix:
        case '1':
            histogramme()
        case '2':
            courbeTemporelle()
        case '3':
            boiteAMoustache()
        case '4':
            heatMap()
        case 'q' :
            break
        case _:
            print("CHOIX INVALIDE")  
        
```
